# 문자열 덧셈 하기 - 문자열 조작 및 반복/조건문 활용

**문제 풀기 전에 어떻게 풀어야할지 생각부터 해봅시다!**

> 사람은 덧셈을 할때 뒤에서부터 계산하고, 받아올림을 합니다.
> 
> 문자열 2개를 받아 덧셈을 하여 숫자를 반환하는 함수 `my_sum(num1, num2)`을 만들어보세요.



**절대로 `return int(num1)+int(num2)` 이렇게 풀지 맙시다!!**

**재귀함수도 사용할 필요 없습니다.**

```python
# 여기에 코드를 작성하시오.
def my_sum(num1, num2):
    list_num1 = list(num1)
    list_num2 = list(num2)
    list_num3 = []
    min_num = min(num1, num2)
    max_num = max(num1, num2)
    string = ''
    i = -1
    
    for n in range(len(max_num)+1):
        list_num3.append(0)
    
    if len(max_num) == len(min_num):
        for n in range(len(min_num)):
            list_num3[i] += (int(list_num1[i]) + int(list_num2[i]))
            if list_num3[i] >= 10:
                list_num3[i] -= 10
                list_num3[i-1] += 1
            i -= 1
    elif len(num1) > len(num2):
        for n in range(len(min_num)-1):
            list_num3[i] += int(list_num1[i])
            list_num3[i] += int(list_num2[i])
            if list_num3[i] >= 10:
                list_num3[i] -= 10
                list_num3[i-1] += 1
            i -= 1
        for n in range(len(max_num)-len(min_num)+1,-1,-1):
            list_num3[n] += int(min_num[n])
            if list_num3[n] >= 10:
                list_num3[n] -= 10
                list_num3[n-1] += 1
    elif len(num1) < len(num2):
        for n in range(len(min_num)):
            list_num3[i] += int(list_num1[i])
            list_num3[i] += int(list_num2[i])
            if list_num3[i] >= 10:
                list_num3[i] -= 10
                list_num3[i-1] += 1
            i -= 1
        for n in range(len(max_num)-len(min_num),0,-1):
            list_num3[n] += int(max_num[n])
            if list_num3[n] >= 10:
                list_num3[n] -= 10
                list_num3[n-1] += 1
                
    if list_num3[0] == 0 :
        del list_num3[0]
        
    for n in range(len(list_num3)):
        string += str(list_num3[n])
        
    return string
```
